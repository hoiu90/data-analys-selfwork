```python
# an example inspired on: https://matplotlib.org/stable/gallery/mplot3d/voxels.html#sphx-glr-gallery-mplot3d-voxels-py

import matplotlib.pyplot as plt

# array of shape 3 x 3 x 3 initialization
# coordinates (x, z, y)
x = np.ones((3, 3, 3))

# On the right, we filter the array x using the condition < 3, which returns
# an array of the same dimensions, where values that satisfy the condition are
# stored as the boolean value True, and those that do not satisfy it are stored as False
# although there are several ways to perform this conversion

cube1 = (x < 3)

# printing data of an array
display(cube1)

# you can test multiple examples - one at a time
# cube1[:,:,1] = False
# cube1[1,:,:] = False
cube1[:,1,:] = False
print(cube1)

colors = np.empty(cube1.shape, dtype=object)
colors[cube1] = 'green'

ax = plt.figure().add_subplot(projection='3d')
ax.axes.set_xlim3d(left=0, right=5)
ax.axes.set_ylim3d(bottom=0, top=5)
ax.axes.set_zlim3d(bottom=0, top=5)
ax.voxels(cube1, facecolors=colors, edgecolor='k')


plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# NumPy offers many data types that can be thought of as containers for information in NumPy arrays.
# These types do not correspond exactly to the types we know from the standard Python library, since “under the hood”
# we still have a C implementation.
# I will present a few examples here, but the number of available types is extensive, so
# I refer you to the documentation for a complete reference.

# As you have likely already noticed, a data type is assigned to an array automatically if we do not specify it
# during its initialization. However, this will not always be optimal in terms of memory utilization for
# the data in that array.

def print_details(np_arr):
    print(np_arr, np_arr.dtype, np_arr.itemsize, np_arr.nbytes)

# we can define data type during array creation
arr_int = np.arange(1, 11, dtype='i') # int32 or int64 in most cases
arr_int2 = np.arange(1, 11, dtype='i4') # np.int32
arr_int3 = np.arange(1, 11, dtype='i1') # np.int8
arr_int4 = np.arange(1, 11, dtype='i2') # np.int16
arr_int5 = np.arange(1, 11, dtype='i8') # np.int64

print_details(arr_int)
print_details(arr_int2)
print_details(arr_int3)
print_details(arr_int4)
print_details(arr_int5)
```

```python
# Each of these types also has an unsigned counterpart—that is, unlike the examples in the previous listing,
# integers are stored as unsigned, which doubles the range of decimal values that can be assigned to such a variable

arr_uint1 = np.arange(1, 11, dtype='u4') # lub np.int32
arr_uint2 = np.arange(1, 11, dtype='u8') # lub np.int64

print_details(arr_uint1)
print_details(arr_uint2)
```

```python
# floating-point types
# float - f8 - 8 bytes (same as float64)
# float16 - f2 - 2 bytes
# float32 - f - 4 bytes
# float64 - d - 8 bytes

arr_1 = np.arange(1, 11, dtype='float16')
arr_2 = np.arange(1, 11, dtype='float32')
arr_3 = np.arange(1, 11, dtype='float64')

print_details(arr_1)
print_details(arr_2)
print_details(arr_3)
```

```python
# other data types

# bool
bool_arr = np.arange(1, 11, dtype='b')
# string - unicode
unknown_arr = np.array(['Ala', 'ma', 'kota'])
unknown_arr2 = np.array(list('What could it be?'))


print_details(bool_arr)
print_details(unknown_arr)
print_details(unknown_arr2)
```

2.3.2 Numpy broadcasting
Let's consider an example of running a function on each element of an ordinary Python list.

```python
my_list = list(range(1, 11))
print(my_list)

for idx in range(len(my_list)):
    my_list[idx] = my_list[idx] * 2

print(my_list)

# or in a more Pythonic way

my_list = [elem * 2 for elem in my_list]
my_list
```

```python
# numpy works in a different manner

arr = np.arange(1,11)
print(arr)
print(arr * 2)
# The operation is performed element-wise, i.e., on each element of the array
# or, in other words, it is broadcasted to each element of the array
```

```python
# Most functions in the NumPy library also accept an array as an input argument
# and perform an operation on each of its elements
# Below, we will calculate the square root of each value in the array
np.sqrt(arr)
```


2.3.3 Indexing and slicing

```python
# indexing on a numpy array works in the same way as other index based collections
arr = np.arange(1, 11)
print(arr[0])

print(arr[-1])

print(arr[:5])
```

```python

# slicing a multidimensional arrays is similar, but we need to provide an index or slice for each dimension

arr = np.arange(1, 17).reshape((4,4))
print(arr)

print(arr[0,0])
print(arr[:,0]) # all "rows", first "column"
print(arr[0,:]) # first "row"
print(arr[0:2,0:2]) # a 2x2 array
```

```python
# we can also assign a value to a slice from an array

arr[:,0] = 0
arr
```

If we now recall the example of an array that was subsequently displayed in a graph, we were dealing with a **boolean array**, which is also called a **boolean map**. This name refers to the operations that can be performed on other arrays using such arrays. Such a map allows you to extract, assign, or apply a function to every value in the array to which we "apply" the map—that is, the map will perform the operation on every logical True value in the array and skip the operation for every False value. Examples follow.

```python
arr = np.arange(1, 17).reshape((4,4))
print(arr)

# simple condition - returns a map where the key has the same dimensions and takes the value True wherever the condition is met
print((arr < 10))

# now we apply a map to an original array
print(arr[(arr < 10)])

# this works like a filter
# other examples will be presented on our future lectures
```

2.4 Examples on how to compare computation time
We already know that familiarity with the NumPy library is necessary to work efficiently with the Pandas library, which is extremely popular in the context of data analysis. Another important aspect is the processing speed offered by NumPy, especially when compared to Python’s standard library and equivalent calculations.

Below are several examples comparing the performance of various Python code snippets using the timeit library.

```python
from timeit import timeit


setup= """
import numpy as np
"""

setup_2 = """
from array import array
"""

stmt1 = """list(range(1, 100001))"""
stmt2 = """np.arange(1, 100001)"""
stmt3 = """array('I', range(1, 100001))"""

timeit(stmt=stmt1, number=1000), timeit(stmt=stmt2, setup=setup, number=1000), timeit(stmt=stmt3, setup=setup_2, number=1000)
```

As shown above, you can see the difference in speed between standard lists and a NumPy array in this specific case, where we create an array of 100,000 integer values. The entire process is repeated 1,000 times to average out the results.
