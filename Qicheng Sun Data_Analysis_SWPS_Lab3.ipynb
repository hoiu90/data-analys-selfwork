````python
# Solutions for tasks 1-9
import numpy as np
import timeit

# 1) 8x8 array with values 1..64
a_small_one = np.arange(1, 65).reshape(8, 8)
print("a_small_one:\n", a_small_one)
````

task 2.For the array a_small_one, calculate:

* the sum for each column,
* the cumulative sum for each row,
* the average value for each row,

````python
# 2) Basic calculations on a_small_one
sum_each_column = a_small_one.sum(axis=0)
cumsum_each_row = a_small_one.cumsum(axis=1)
avg_each_row = a_small_one.mean(axis=1)

print("\nSum for each column:\n", sum_each_column)
print("\nCumulative sum for each row:\n", cumsum_each_row)
print("\nAverage value for each row:\n", avg_each_row)
````

task 3.Return a subarray from the array a_small_one containing even-numbered elements (do not overwrite this variable).

````python
# 3) Subarray with even elements (without overwriting a_small_one)
even_subarray = a_small_one[a_small_one % 2 == 0]
print("\nEven elements subarray:\n", even_subarray)
````

task 4.Save the array a_small_one to a variable named no_even_small_one, replacing even values with a value such as nan.



````python
# 4) Replace even values with nan in a copy
no_even_small_one = a_small_one.astype(float).copy()
no_even_small_one[no_even_small_one % 2 == 0] = np.nan
print("\nno_even_small_one (even -> nan):\n", no_even_small_one)
````

task 5.For the array no_even_small_one, calculate the following using functions that handle nan values (these functions usually have the prefix nan):

* the sum for each row,
* the average for each column,
* the smallest value in the entire array,

````python
# 5) NaN-aware calculations
nan_sum_each_row = np.nansum(no_even_small_one, axis=1)
nan_avg_each_column = np.nanmean(no_even_small_one, axis=0)
nan_min_all = np.nanmin(no_even_small_one)

print("\nNaN-aware row sums:\n", nan_sum_each_row)
print("\nNaN-aware column means:\n", nan_avg_each_column)
print("\nNaN-aware minimum value in array:\n", nan_min_all)
````

task 6.Insert the following into the no_even_small_one array using the np.where function:

* the value 1 for each np.nan value (see the built-in np.isnan() function)
* even numbers, which we previously replaced with nan values (now 1 if the first step is completed), so we restore the array to its original state from the a_small_one array—for the condition, insert the value from a_small_one or no_even_small_one

````python
# 6) Use np.where:
#    a) replace each nan by 1
no_even_small_one = np.where(np.isnan(no_even_small_one), 1, no_even_small_one)
print("\nAfter replacing nan with 1:\n", no_even_small_one)

#    b) restore original even numbers from a_small_one
#       (positions that were set to 1 correspond to former even values)
no_even_small_one = np.where(no_even_small_one == 1, a_small_one, no_even_small_one)
print("\nAfter restoring even numbers:\n", no_even_small_one)
````

task 7.Create an array named normal_100 with dimensions 100x100 by sampling values from a standard normal distribution using a np.random module.

````python
# 7) 100x100 array from standard normal distribution
rng = np.random.default_rng()
normal_100 = rng.standard_normal((100, 100))
````

task 8.For the normal_100 array, calculate:

* the mean value in each row and column (separately),
* calculate the average absolute difference for the values in the previous point relative to the value 0 (this is the expected value given our assumptions about the normal distribution) — separately for the column averages and the row averages; keep in mind that operations on a NumPy array are performed in broadcast mode.

````python
# 8) Means and average absolute difference from 0
row_means = normal_100.mean(axis=1)
col_means = normal_100.mean(axis=0)

avg_abs_diff_row_means = np.mean(np.abs(row_means - 0))
avg_abs_diff_col_means = np.mean(np.abs(col_means - 0))

print("\nnormal_100 row means (first 10):\n", row_means[:10])
print("\nnormal_100 column means (first 10):\n", col_means[:10])
print("\nAverage absolute difference from 0 (row means):", avg_abs_diff_row_means)
print("Average absolute difference from 0 (column means):", avg_abs_diff_col_means)
````

task 9. Create a 1000x1000 array by filling it with random values from a standard normal distribution, and then, using examples from previous labs (the timeit module), compare the execution times for this array:

* writing to a text file vs. writing to a file in pickle format vs. writing to a compressed file (the savez_compressed function in the NumPy library documentation)
* reading arrays back from the three files mentioned above.

````python
# 9) Time comparison for save/load methods on a 1000x1000 normal array
big_normal = rng.standard_normal((1000, 1000))

# Save methods
save_txt_time = timeit.timeit(
    stmt="np.savetxt('big_normal.txt', big_normal)",
    globals=globals(),
    number=1,
)

save_npy_time = timeit.timeit(
    stmt="np.save('big_normal.npy', big_normal)",
    globals=globals(),
    number=3,
)

save_npz_time = timeit.timeit(
    stmt="np.savez_compressed('big_normal_compressed.npz', big_normal)",
    globals=globals(),
    number=3,
)

# Load methods
load_txt_time = timeit.timeit(
    stmt="_ = np.loadtxt('big_normal.txt')",
    globals=globals(),
    number=1,
)

load_npy_time = timeit.timeit(
    stmt="_ = np.load('big_normal.npy')",
    globals=globals(),
    number=10,
)

load_npz_time = timeit.timeit(
    stmt="_ = np.load('big_normal_compressed.npz')['arr_0']",
    globals=globals(),
    number=10,
)

print("\n--- Time comparison (seconds) ---")
print(f"Save TXT (1 run): {save_txt_time:.4f}")
print(f"Save NPY (3 runs): {save_npy_time:.4f}")
print(f"Save NPZ compressed (3 runs): {save_npz_time:.4f}")
print(f"Load TXT (1 run): {load_txt_time:.4f}")
print(f"Load NPY (10 runs): {load_npy_time:.4f}")
print(f"Load NPZ compressed (10 runs): {load_npz_time:.4f}")
````
